1. Define a function

    $f(x)=e^x$
    

2. Define a domain

    $-3 < x < 3
    

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(-3, 3, 0.01)
y = np.exp(x)
```

3. Plot the function


```python
plt.plot(x, y, label = 'f(x)')
```

4. Select a point

    $ a = 0
    

```python
a = 0
```

5. Mark the point


```python
plt.plot(x, y, label = 'f(x)')
plt.plot(a, np.exp(a), 'o')
```

6. Define the first-order Taylor approximation at the point

    $g(x) = e^0 + e^0(x - 0) = x + 1$
    

```python
y2 = np.exp(a) + np.exp(a) * (x - a)
```

7. Plot the Talyor approximation


```python
plt.plot(x, y, label = 'f(x)')
plt.plot(a, np.exp(a), 'o')
plt.plot(x, y2, label = 'g(x)')

plt.title('Taylor Approximation')
plt.xlabel('x axis')
plt.ylabel('y axis')

plt.legend()
plt.show()
```
